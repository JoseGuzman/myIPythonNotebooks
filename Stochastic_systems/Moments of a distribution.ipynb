<H1>Moments of a distribution</H1>

```python
%pylab inline
import scipy.stats as stats
```

```python
mu, sigma = 0.0, 1.0
data = np.random.normal(mu, sigma, 10000)
```

```python
plt.hist(data, normed=1,bins=20, facecolor='yellow', alpha=.5);

xval = np.linspace(-4, 4, 100)
rv = stats.norm(mu, sigma)
plt.plot(xval,rv.pdf(xval), color='red');
```

```python
np.mean(data)
```

```python
np.var(data)
```

```python
stats.skew(data)
```

```python
stats.kurtosis(data)
```
