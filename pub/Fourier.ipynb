```python
A = 100 # amplitude
dt = 1/6. # sampling in hours
f = 2.0 # 2 per hours

t = np.arange(0,24, dt)

signal = A*np.sin(2*np.pi*f*t)

fig = figure(figsize=(12,7))
plt.plot(t,signal)
```
