<H1> Correlation of discrete signals </H1>

Correlation is a measurement of how similar two signals are
$Coo

```python
a = np.array( [0.1, 0.2, -0.1, 4.1, -2.0, 1.5, -0.1 ]);
b = np.array( [0.1, 4.0, -2.2, 1.6, 0.1,  0.1, 0.2]) ;
plt.plot(a)
plt.plot(b)
```

<B>Cross correlation:</B> measure of similiarity between two signals at different lag positions.

```python
len(a)==len(b)
```

```python
# Calculate correlation at zero lag
# multiply the vertically aligned samples and sum them all
np.sum(a*b)
```

```python
# Calculate correlation at a lag of one
np.sum(a[1:]*b[:-1])
```

```python
# Calculate correlation at a lag of two
np.sum(a[2:]*b[:-2])
# that large value indicates that there's a large similarity between the two signals at that overlap position
```

```python
# To compute the correlation sequence we just continue through all lags
N = len(a)
corr_seq = list([np.sum(a*b)])
for i in range(1,N):
    corr_seq.append( np.sum(a[i:]*b[:-i]) )
print(corr_seq)
```

```python
# we need to compute the negative lag
corr_seq_n = list()
for i in range(1,N):
    corr_seq_n.append( np.sum(b[i:]*a[:-i]) )
print(corr_seq_n)
```

```python
ycorr = np.correlate(a,b,'full')
N = len(a)-1
lag = np.arange(-N, N+1)
print(lag, ycorr)
```

```python
plt.plot(lag, ycorr)
```
