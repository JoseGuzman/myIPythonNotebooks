<H2> The Gaussian function </H2>
A Gaussian function is a function with a symmetric bell shape. It has the following form:

$f(x) = a \exp{\left(- { \frac{(x-b)^2 }{ 2 c^2} } \right)}, $

where $a$ is the height of the curve, $b$ is the position at the center of the peak and $c$ is the width of the curve.

```python
from scipy.optimize import curve_fit
from math import exp

@np.vectorize
def fgauss(x, a, b, c):
    """
    simple gaussian function
    """

    val = a*exp( -( (x-b)**2/(2*c**2) ) )
    return(val)

# plot example
dt = np.arange(-200,200, 0.1) # independent variable
y = fgauss(dt, a=70, b=0, c=58) 
plt.plot(dt, y, color='#AA0000', lw=1.5);

plt.xlim(xmax=200, xmin=-200);
plt.ylim(ymax=100);
plt.xlabel('$\Delta$ t (post-pre)');
plt.ylabel('LTP magnitude (% of baseline)');
plt.title('Example Gaussian function');
```

<H2> Fitting electrophysiology data </H2>

```python
LTP = np.array([1.2187022977, 1.311189796, 1.522211156, 1.4848593141, 1.501551, 1.7778987096, 1.401036303, 1.4136346933, 1.14105007
])
DeltaT = np.array([-100, -50, -20, -10, 0, 10, 20, 50, 100], dtype=float)
#yerr = np.array([8.7501216923, 17.8016333, 25.875, 12.3952680338, 16.0634901029, 25.5986376591, 17.9779865391, 19.5, 6.55615131])


```

```python
# perform fitting
myparam = curve_fit(f=fgauss, xdata =DeltaT , ydata =LTP, p0=(70, 0, 58))
a, b, c = myparam[0]
print("a = %2.4f, b = %2.4f, c= %f"%(a,b, c))
```

```python
# Plot everything together
dt = np.arange(-200,200, 0.1)
yfit =fgauss(dt, a, b, c)

plt.plot(dt, yfit, color='#AA0000');
plt.plot(DeltaT, LTP, 'ko', markersize= 6, color='#969696')
#plt.errorbar(DeltaT, LTP, yerr=yerr, fmt='ko', markersize=8, lw=1.5);

plt.xlim(xmax=200, xmin=-200);
#plt.ylim(ymax=125);
plt.xlabel('$\Delta$ t (post-pre)');
plt.ylabel('LTP magnitude (% of baseline)');
```
