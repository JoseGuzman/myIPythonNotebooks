<H1>Deconvolution example</H1>

```python
%pylab inline
from scipy.signal import deconvolve
```

<H2>Trace 2</H2>

```python
!ls
```

```python
# load trace
trace = np.load('IPSCs2.npy')
plt.plot(trace)
plt.ylim(-85, -65);
```

<H2>Select a template<H2>

```python
template = trace[1000:4000]
plt.plot(template)
```

```python
quotient, remainder = deconvolve(signal = trace, divisor=template)
```

```python
plt.plot(quotient)
```
