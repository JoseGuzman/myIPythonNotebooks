<H1>Deconvolution example</H1>

```python
%pylab inline
from scipy.signal import deconvolve
```

```python
# load trace
trace = np.load('IPSCs.npy')
plt.plot(trace)
plt.ylim(-85, -65);
```

<H2>Select a template<H2>

```python
template = trace[1000:4000]
plt.plot(template)
```

```python
plt.plot(trace[4000:])
```

```python
quotient, remainder = deconvolve(signal = trace, divisor=template)
```

```python
plt.plot(quotient)
```

```python
plt.plot(remainder)
```
